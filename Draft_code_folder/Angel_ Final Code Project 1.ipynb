```python
import wbdata
import matplotlib.pyplot as plt
%matplotlib widget

indicators = {"SH.DYN.AIDS.ZS":'HIV Prevalence in Botswana',
              "SP.DYN.TFRT.IN":'Total Fertility Rate in Botswana'}
country_code = "BWA"

data_bot = wbdata.get_dataframe(indicators=indicators, country=country_code).dropna()

data_bot = data_bot.sort_index()
```

```python
fig,ax = plt.subplots()

data_bot.reset_index().plot(x='date', y='HIV Prevalence in Botswana', ax=ax, color='red', label='HIV Prevalence in Botswana')

ax2 = ax.twinx()

data_bot.reset_index().plot(x='date', y='Total Fertility Rate in Botswana', ax=ax2, color='purple', label='Total Fertility Rate in Botswana')

ax.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```
